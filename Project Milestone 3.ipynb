```python
import pandas as pd

df = pd.read_html('https://www.baseball-reference.com/teams/')[0]

df.head()
```

## Transformation 1: Removing unneeded rows

As shown above, there are multiple rows in this dataset that are unneeded. The rows that relate to "also known as" or "also played as" exist in the table but do not contain any relevant datapoints about the team. To handle this, we will remove all rows that contain "NaN" in the "Rk" column.

```python
df = df.dropna(subset=['Rk'])
df.head(10)
```

## Transformation 2: Adding short name column

This dataset does not include values for the teams "short names" or their abbreviations. To remedy this, I'll create a dictionary to reference and then map each value to it's "franchise" value to its associated short name. This will help to join datasets later on via the short name column. 

```python
mlb_teams = {
    'Arizona Diamondbacks': 'ARI',
    'Atlanta Braves': 'ATL',
    'Baltimore Orioles': 'BAL',
    'Boston Red Sox': 'BOS',
    'Chicago Cubs': 'CHC',
    'Chicago White Sox': 'CWS',
    'Cincinnati Reds': 'CIN',
    'Cleveland Guardians': 'CLE',
    'Colorado Rockies': 'COL',
    'Detroit Tigers': 'DET',
    'Houston Astros': 'HOU',
    'Kansas City Royals': 'KC',
    'Los Angeles Angels': 'LAA',
    'Los Angeles Dodgers': 'LAD',
    'Miami Marlins': 'MIA',
    'Milwaukee Brewers': 'MIL',
    'Minnesota Twins': 'MIN',
    'New York Mets': 'NYM',
    'New York Yankees': 'NYY',
    'Oakland Athletics': 'OAK',
    'Philadelphia Phillies': 'PHI',
    'Pittsburgh Pirates': 'PIT',
    'San Diego Padres': 'SD',
    'Seattle Mariners': 'SEA',
    'San Francisco Giants': 'SF',
    'St. Louis Cardinals': 'STL',
    'Tampa Bay Rays': 'TB',
    'Texas Rangers': 'TEX',
    'Toronto Blue Jays': 'TOR',
    'Washington Nationals': 'WSH'
}

df['Short Name'] = df['Franchise'].map(mlb_teams)

df.head()
```

## Transformation 3: Renaming columns

Some of the columns in the dataset are not totally clear as to their meaning. So, I'm renaming them to ensure clarity during the final analysis. 

```python
df = df.rename(
    columns={
    'Pnnts': 'Pennants',
    'Divs': 'Division Titles',
    'Franchise': 'Team Name',
    'W': 'Wins',
    'L': 'Losses',
    'G': 'Games Played',
    'R': 'Runs Scored',
    'AB': 'At Bats',
    'H': 'Hits',
    'BA': 'Batting Average',
    'RA': 'Runs Allowed'
    }
)

df.head()
```

## Transformation 4: Creating Years as a Team column

I want to create some columns that are based on how long the team has existed (for example, runs allowed per year, hits per year, runs scored per year, etc.) To do this, I first need to create a column that contains how many years the team has existed. As of now, the "from" and "to" columns, are stored as text, so they must be converted to integers using the .to_numeric function. Then the column can be created using simple math. This is shown below. 

```python
df['From'] = pd.to_numeric(df['From'])
df['To'] = pd.to_numeric(df['To'])

df['Years as a team'] = df['To'] - df['From']

df.head()
```

## Transformation 5: Creating "Per Year" columns

Now that we know how many years each team has existed, we can create the aforementioned "per year" columns. This will allow us to compare how effective/ineffective a team has been over team while taking into account that some teams have existed for much longer than others. 

```python
df['Years as a team'] = pd.to_numeric(df['Years as a team'])
df['Runs Scored'] = pd.to_numeric(df['Runs Scored'])
df['Runs Allowed'] = pd.to_numeric(df['Runs Allowed'])
df['HR'] = pd.to_numeric(df['HR'])
df['Hits'] = pd.to_numeric(df['Hits'])

df['Runs Scored Per Year'] = df['Runs Scored'] / df['Years as a team']
df['Runs Allowed Per Year'] = df['Runs Allowed'] / df['Years as a team']
df['Home Runs Per Year'] = df['HR'] / df['Years as a team']
df['Hits Per Year'] = df['Hits'] / df['Years as a team']

df.head(50)
```
