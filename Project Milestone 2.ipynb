```python
import pandas as pd

df = pd.read_csv("C:/Users/DarinPC/Desktop/Bellevue/DSC 540/ballparks.csv")
df.head()
```

Step 1. Renaming columns for clarity. Only small changes just to ensure easy readability. 

```python
df = df.rename(
    columns={
    'team_name': 'team_short_name', 
    'ballpark': 'ballpark_name', 
    'left_field': 'left_field_distance(ft)',
    'center_field': 'center_field_distance(ft)',
    'right_field': 'right_field_distance(ft)',
    'elevation': 'elevation (ft)',
    'roof': 'roof_on_%',
    'daytime': 'daytime_games_%'
}
            )
df.head()
```

Step 2. Categorize ballparks into temperature categories. This uses a user defined function and then passes values from the 'avg_temp' column into the function via the apply method to categorize each value, and then creates a new column based on the results. 

I plan to use this new column to group stadiums together to see if there are any noticable effects as it relates to temperature categories. 

```python
def categorize_temperature(temp):
    if temp < 70:
        return 'Cold'
    elif 70 <= temp < 75:
        return 'Moderate'
    else:
        return 'Hot'

df['temperature_category'] = df['avg_temp'].apply(categorize_temperature)

df.head()
```

Step 3. Categorize ballparks based on time spent played during the day/night. This uses a user defined function and then passes values from the 'daytime' column into the function via the apply method to categorize each value, and then creates a new column based on the results. 

I plan to use this new column to see if teams perform better/worse depending on their tendency in playing during the day/night. 

```python
def categorize_time_of_day(daytime_percent):
    if daytime_percent < 0.35:
        return 'More Night Games'
    elif 0.35 <= daytime_percent < .45:
        return 'Typical Mix'
    else:
        return 'More Day Games'

df['time_of_day_category'] = df['daytime_games_%'].apply(categorize_time_of_day)

df.head()
```

Step 4. Create 1/0 flag whether a ballpark is home run friendly or not, based on hr_park_effects (which is a measure of total "park effects" on home run rate, with 100 being a baseline average).

I plan to use this column to see how a park being "home run friendly" impacts success. 

```python
df['home_run_friendly'] = df['hr_park_effects'].apply(lambda x: 1 if x > 100 else 0)

df.head()
```

Step 5. Create a total distance measure that sums all outfield wall distances. 

I plan to use this column to see if stadiums with longer outfield walls (in total) experience more/less success. 

```python
df['total_wall_distance(ft)'] = df['left_field_distance(ft)'] + df['center_field_distance(ft)'] + df['right_field_distance(ft)']

df.head()
```
